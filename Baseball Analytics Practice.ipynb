```python
import pandas as pd
import numpy as np
```

```python
Pitching = pd.read_csv(r'stats_mlb_savant_pitching_updated.csv')
```

```python
print (Pitching.head())
```

```python
# library
import seaborn as sns
import matplotlib.pyplot as plt


# A scatterplot without regression fit
ax = sns.regplot(x=Pitching["FIP"], y=Pitching["fastball_avg_speed"], fit_reg=False)

plt.show()
```

```python
sns.regplot(x=Pitching["FIP"], y=Pitching["fastball_avg_speed"])

# Titles
plt.xlabel('FIP')
plt.ylabel('Fastball Avg Speed (mph)')
plt.title('Does Velocity Equal Effectiveness?', loc='center')
plt.show()
```

```python

plt.plot( 'FIP', 'fastball_avg_speed', "", data=Pitching, linestyle='', marker='o', markersize=3, alpha=0.55, color="purple")
 
plt.xlabel('FIP')
plt.ylabel('Fastball Avg Speed (mph)')
plt.title('Does Velocity Equal Effectiveness?', loc='center')
plt.show()
```
